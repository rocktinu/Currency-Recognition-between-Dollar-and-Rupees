```python
import os
import warnings
warnings.simplefilter('ignore')
```

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from skimage.io import imread, imshow
from skimage.transform import resize
from skimage.color import rgb2gray
```

```python
dollar=os.listdir("C:/Users/HARSH UPADHYAY/Desktop/harsh_image_recognition/dollor_data_set")
```

```python
rupees=os.listdir("C:/Users/HARSH UPADHYAY/Desktop/harsh_image_recognition/rupees_data_set")
```

```python
limit=100
dollar_images=[None]*limit
j=0
for i in dollar:
    if (j<limit):
        dollar_images[j]=imread("C:/Users/HARSH UPADHYAY/Desktop/harsh_image_recognition/dollor_data_set/"+i)
        j+=1
    else:
        break
```

```python
limit=100
rupees_images=[None]*limit
j=0
for i in rupees:
    if (j<limit):
        rupees_images[j]=imread("C:/Users/HARSH UPADHYAY/Desktop/harsh_image_recognition/rupees_data_set/"+i)
        j+=1
    else:
        break
```

```python
dollar_gray=[None]*limit
j=0
for i in dollar:
    if(j<limit):
        dollar_gray[j]=rgb2gray(dollar_images[j])
        j+=1
    else:
        break
```

```python
rupees_gray=[None]*limit
j=0
for i in rupees:
    if(j<limit):
        rupees_gray[j]=rgb2gray(rupees_images[j])
        j+=1
    else:
        break
```

```python
imshow(dollar_gray[1])
```

```python
imshow(rupees_gray[1])
```

```python
for j in range (100):
    dl=dollar_gray[j]
    dollar_gray[j]=resize(dl,(64,64))
```

```python
for j in range (100):
    rs=rupees_gray[j]
    rupees_gray[j]=resize(rs,(64,64))
```

```python
len_of_images_dollar=len(dollar_gray)
```

```python
image_size_dollar=dollar_gray[1].shape
```

```python
flatten_size_dollar=image_size_dollar[0]*image_size_dollar[1]
```

```python
flatten_size_dollar
```

```python
for i in range(len_of_images_dollar):
    dollar_gray[i]=np.ndarray.flatten(dollar_gray[i]).reshape(flatten_size_dollar,1)
```

```python
dollar_gray=np.dstack(dollar_gray)
```

```python
dollar_gray=np.rollaxis(dollar_gray,axis=2,start=0)
```

```python
dollar_gray.shape
```

```python
dollar_gray=dollar_gray.reshape(len_of_images_dollar,flatten_size_dollar)
```

```python
dollar_gray.shape
```

```python
dollar_data=pd.DataFrame(dollar_gray)
```

```python
dollar_data
```

```python
dollar_data["label"]="dollar"
```

```python
dollar_data
```

```python
len_of_images_rupees=len(rupees_gray)
```

```python
image_size_rupees=rupees_gray[1].shape
```

```python
flatten_size_rupees=image_size_rupees[0]*image_size_rupees[1]
```

```python
flatten_size_rupees
```

```python
for i in range(len_of_images_rupees):
    rupees_gray[i]=np.ndarray.flatten(rupees_gray[i]).reshape(flatten_size_rupees,1)
```

```python
rupees_gray=np.dstack(rupees_gray)
```

```python
rupees_gray=np.rollaxis(rupees_gray,axis=2,start=0)
```

```python
rupees_gray=rupees_gray.reshape(len_of_images_rupees,flatten_size_rupees)
```

```python
rupees_gray.shape
```

```python
rupees_data=pd.DataFrame(rupees_gray)
```

```python
rupees_data
```

```python
rupees_data["label"]="rupees"
```

```python
rupees_data
```

```python
currency=pd.concat([dollar_data,rupees_data])
```

```python
currency
```

```python
from sklearn.utils import shuffle
```

```python
currency_indexed=shuffle(currency).reset_index()
```

```python
currency_indexed
```

```python
currency_data=currency_indexed.drop(['index'],axis=1)
```

```python
currency_data
```

```python
currency_data.to_csv("C:/Users/HARSH UPADHYAY/Desktop/harsh_image_recognition/currency.csv")
```

```python
x=currency_data.values[:,:-1]
```

```python
y=currency_data.values[:,-1]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn import svm
```

```python
clf = svm.SVC()
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
for i in (np.random.randint(0,10,5)):
    predicted_images = (np.reshape(x_test[i], (64,64)).astype(np.float64))
    plt.title('predicted label: {0}'. format(y_pred[i]))
    plt.imshow(predicted_images, interpolation='nearest', cmap='gray')
    plt.show()
```

```python
from sklearn import metrics
```

```python
accuracy=metrics.accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
from sklearn.metrics import confusion_matrix
```
